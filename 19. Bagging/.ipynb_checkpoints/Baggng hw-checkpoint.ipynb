```python
df.describe()
```

```python
df1 = df[df.Cholesterol<=(df.Cholesterol.mean()+3*df.Cholesterol.std())]
df1.shape
```

```python
df1[df1.Cholesterol>(df1.Cholesterol.mean()+3*df1.Cholesterol.std())]
```

```python
df[df.MaxHR>(df.MaxHR.mean()+3*df.MaxHR.std())]
```

```python
df[df.FastingBS>(df.FastingBS.mean()+3*df.FastingBS.std())]
```

```python
df2 = df1[df1.Oldpeak<=(df1.Oldpeak.mean()+3*df1.Oldpeak.std())]
df2.head()
df2.shape
```

```python
df3 = df2[df2.RestingBP<=(df2.RestingBP.mean()+3*df2.RestingBP.std())]
df3.shape
```

```python
df.ChestPainType.unique()
```

```python
df.RestingECG.unique()
```

```python
df.ExerciseAngina.unique()
```

```python
df.ST_Slope.unique()
```

```python
df4 = df3.copy()
df4.RestingECG.replace({'Normal': 0, 'ST':1, 'LVH':2  }, inplace =True)
df4.ExerciseAngina.replace({'N': 0, 'Y':1  }, inplace =True)
df4.ST_Slope.replace({'Up': 0, 'Flat':1, 'Down':2  }, inplace =True)
df4.head()
```

```python
df5 = pd.get_dummies(df4, drop_first = True)
df5.head()
```

```python
X = df5.drop('HeartDisease', axis =1)
y = df5.HeartDisease

X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_scaled = sc.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2,random_state=20)
```

```python
X_train.shape
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

scores= cross_val_score(SVC(), X_scaled, y, cv = 5)
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bm = BaggingClassifier(base_estimator=SVC(), n_estimators=100, max_samples=0.8, random_state=0 )
scores = cross_val_score(bm, X_scaled, y, cv=5)
scores.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
score = cross_val_score(DecisionTreeClassifier(), X_scaled, y, cv=5)
score.mean()
```

```python
bm1 = BaggingClassifier(base_estimator=DecisionTreeClassifier(random_state=0), n_estimators=100, max_samples=0.9,oob_score=True, random_state=0 )
scores= cross_val_score(bm1, X_scaled, y, cv=5)
scores.mean()
```
